```python
# Mendeteksi nilai NA yang diisi secara manual, seperti NA, N/A, na, atau n/ a
missing_values = ["n/a", "na",  "--"]
```

```python
df = pd.read_csv("assets/property_data.csv", na_values = missing_values)
```

```python
df["NUM_BEDROOMS"]
```

```python
# Mengisi Ruang Kosong (NA) dengan menggunakan Pandas
df["NUM_BEDROOMS"].isnull()
```

```python
df["OWN_OCCUPIED"]
```

```python
df["OWN_OCCUPIED"].isnull()
```

```python
df.head(9)
```

```python
cnt=0
for row in df["OWN_OCCUPIED"]:
    try:
        int(row)
        df.loc[cnt, "OWN_OCCUPIED"] = np.nan
    except ValueError:
        pass
    cnt += 1
```

```python
df.head(9)
```

## Summarizing Missing Values

```python
df.isnull().sum()
```

```python
# Mengecek nilai hilang dari seluruh data dengan cepat
df.isnull().values.any()
```

## Replacing

```python
# Mengisi nilai missing values dengan nilai tunggal

df["ST_NUM"].fillna(125, inplace=True)
```

```python
df.loc[2, "ST_NUM"] = 125
```

```python
df.head(9)
```

```python
# Cara yang sangat umum untuk mengganti missing values dengan menggunakan nilai tengah (median)

median = df["NUM_BEDROOMS"].median()
df["NUM_BEDROOMS"].fillna(median, inplace=True)
```

```python
data_columns = ["Consumption", "Wind", "Solar", "Wind+Solar"]
opsd_weekly_mean = opsd_daily[data_columns].resample("W").mean()
opsd_weekly_mean.head(3)
```

### Hasil Analisis - IV
Baris pertama di atas, berlabel 2006-01-01, berisi rata-rata dari semua data yang ada dalam time bin 2006-01-01 hingga 2006-01-07. Baris kedua, berlabel 2006-01-08, berisi data rata-rata untuk time bin 2006-01-08 hingga 2006-01-14, dan seterusnya.

### Catatan
Berdasarkan konstruksi, deret waktu mingguan kita memiliki 1/7 poin data sebanyak deret waktu harian. Kita dapat mengonfirmasi ini dengan membandingkan jumlah baris dari dua DataFrame.

```python
print(opsd_daily.shape[0])
print(opsd_weekly_mean.shape[0])
```

Menggambarkan Rangkaian Waktu Solar Harian dan Mingguan dalam Satu Periode Enam Bulan untuk membandingkannya.

```python
start = "2017-01"
end = "2017-06"

opsd_daily.loc[start:end, "Solar"].plot(marker=".", linestyle="-", linewidth = 0.5, figsize = (11, 9))
opsd_weekly_mean.loc[start:end, "Solar"].plot(marker="o", linestyle="-", markersize = 8, figsize = (11, 9));
```

Kita dapat melihat bahwa deret waktu rata-rata mingguan lebih halus daripada deret waktu harian karena variabilitas frekuensi yang lebih tinggi telah dirata-ratakan dalam resampling.
